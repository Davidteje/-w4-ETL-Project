### API AMBEE - climatología diaria (últimos 2 días)

```python
import requests
url = "https://api.ambeedata.com/weather/history/daily/by-lat-lng"
querystring = {"lat":"25.3328574490728","lng":"51.4836629324111","from":"2022-11-11 01:00:00","to":"2022-11-12 03:00:00"}
headers = {
    'x-api-key': "01f0c2622b7abfa17f830522b9f0d078da8c42f6bd19167315282f3f3ce7f480",
    'Content-type': "application/json"
    }
response = requests.request("GET", url, headers=headers, params=querystring)
print(response.text)
```

```python
response.json()
```

```python
import pandas as pd
df = pd.DataFrame(response.json())
df
```

```python
df = pd.json_normalize(df.explode('data').data) 
df
```

```python
df = df[:2]
df
```

```python
def conv_int(raw): return int(raw)
```

```python
df.time = df.time.apply(conv_int)
```

```python
df
```

```python
import time
time.localtime(1668269341)
```

```python
for i in df.time:
    print(time.localtime(i))
```

```python
df.time[0] = '2022-11-11'
df.time[1] = '2022-11-12'
df
```

```python
df = df.rename(columns={'time':'date'})
df
```

```python
def conv_fah_to_cent(fah): return ((fah-32) * 5/9)
```

```python
df
```

```python
df.temperature = df.temperature.apply(conv_fah_to_cent)
df
```

```python
df.maxTemperature = df.maxTemperature.apply(conv_fah_to_cent)
```

```python
df.minTemperature = df.minTemperature.apply(conv_fah_to_cent)
df.apparentTemperature = df.apparentTemperature.apply(conv_fah_to_cent)
df.maxApparenttemperature = df.maxApparenttemperature	.apply(conv_fah_to_cent)
df.minApparenttemperature = df.minApparenttemperature.apply(conv_fah_to_cent)
df
```
